```python
data = pd.read_csv("data/raw_data/dataset1_gpt.csv")
pairs = data[["criminal_id", "delito_id"]]

delitos = pairs.groupby('delito_id', as_index=False).agg({"criminal_id":"count"})
delitos = delitos[delitos["criminal_id"]>1]

incidence = pairs[pairs["delito_id"].isin(delitos["delito_id"].unique())].groupby('delito_id')['criminal_id'].apply(list).to_dict()
```

```python
import matplotlib.pyplot as plt
H = hnx.Hypergraph(incidence)
hnx.draw(H,with_edge_labels=False, with_node_labels=False, layout_kwargs={"k":5, "iterations":1000})
print("mi primera visualización de un hipergrafo!! es horrible !!!")
plt.show()
```
